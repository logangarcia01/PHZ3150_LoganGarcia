```python
def srt_all (times, dist, err_dist):
        for i in range(len(times)):
            sort_i = i + np.argmin(times[i:])
            (times[i],times[sort_i]) = (times[sort_i],times[i])
        for j in range(len(dist)):
            sort_j = j + np.argmin(dist[j:])
            (dist[j],dist[sort_j]) = (dist[sort_j],dist[j])
        for k in range(len(err_dist)):
            sort_k = k + np.argmin(err_dist[k:])
            (err_dist[k],err_dist[sort_k]) = (err_dist[sort_k],err_dist[k])
        return times, dist, err_dist
```
